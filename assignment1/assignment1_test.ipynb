# Min-max scaling and euclidean distance

Formula: (x - min) / (max - min)

```python
def normalize(ser: pd.Series):
    val_min = ser.min()
    val_max = ser.max()
    ser_norm = (ser - val_min) / (val_max - val_min)
    return ser_norm

cowdata['age_normalized'] = normalize(cowdata['age'])
cowdata['milk_normalized'] = normalize(cowdata['milk'])

cowdata
```
